```python
from dask import array as darr
import numpy as np
from distributed import Client, performance_report
```

```python
client = Client()
client
```

```python
xy = darr.random.uniform(
    0, 1,
    size=(1e12 / 100, 2), chunks=(500e6 / 50, None)
)
xy
```

```python
pi = 4 * ((xy ** 2).sum(-1) < 1).mean()
pi
```

```python
%%time
with performance_report(filename="dask-perf-report.html"):
    x = pi.compute()
```
